```python
from tkinter import *
from tkinter import ttk
from PIL import ImageTk, Image
from keras.utils import load_img,img_to_array
from keras.models import load_model
import numpy as np
import os
from tkinter.filedialog import * 
from tkinter.messagebox import *


url = 'D:\HKI-2022.2023\AI\Project\Food.csv'
file =''
Label1 = ''
model_CNN = load_model('D:\HKI-2022.2023\AI\Project\Food1.h5')


root = Tk()
root.title('Nhận diện đồ ăn Việt Nam')
root.geometry('650x380')
root.resizable(FALSE,FALSE)
root.iconbitmap('D:\HKI-2022.2023\AI\Project\icon.ico')
root.resizable(False,False)

frame = Frame(root, width=650,height=500,relief=RIDGE,borderwidth=3,bg='WHITE')
frame.place(x=0,y=0)
lb = Label(frame,text = 'NHẬN DIỆN ĐỒ ĂN VIỆT NAM', font = ('Times',18),bg='WHITE', fg='Dark Cyan')
lb.place(x=150, y=10)

frame1 = Frame(frame, width=300,height=250,relief=RIDGE,borderwidth=3,bg='WHITE')
frame1.place(x=0,y=50)

frame2 = Frame(frame, width=345,height=40,relief=RIDGE,borderwidth=3,bg='WHITE')
frame2.place(x=300,y=50)

frame3 = Frame(frame, width=345,height=210,relief=RIDGE,borderwidth=3,bg='WHITE')
frame3.place(x=300,y=90)


my_txt = Message(frame2, font=('Helvetica bold', 12, 'bold'),bg='WHITE',
                 justify=LEFT, text='',
                 width=345)
my_txt.place(x=0,y=5)

en1 = Text(frame3, width=25,height=8, borderwidth=0, relief=RIDGE,font=('Verdana',15))
en1.place(x=0,y=0)



bg_frame1 = Image.open('D:\HKI-2022.2023\AI\Project\icon.jpg')
bg_frame1 = bg_frame1.resize((700,500),Image.Resampling.LANCZOS)
open_bg = ImageTk.PhotoImage(bg_frame1)



def open_file():
	cancel()
	global file,Label1
	file = askopenfilename(defaultextension = '.jpg',filetypes = [('All Files','*.*'),('JPG Files','*.jpg'),('PNG Files','*.png')])
	if file == '':
		showinfo('Thông báo','Vui lòng chọn ảnh!')
	else:
		openfile = Image.open(file)
		openfile = openfile.resize((290,240), Image.Resampling.LANCZOS)
		showfile = ImageTk.PhotoImage(openfile)
		Label1 = Label(frame1,image = showfile)
		Label1.pack()
		mainloop()


def cancel():
	global Label1,file
	if Label1 == '' :
		return
	else:
		Label1.destroy()	
		my_txt.configure(text='')
		en1.delete(1.0,'end')
	file = ''
	

def check():
	global file,model_CNN
	if file =='':
		showinfo('Thông báo','Vui lòng chọn ảnh!')
	else:
		img = load_img(file,target_size =(224,224))
		img = img_to_array(img)
		img = img.astype('float32')
		img = img/255
		img = np.expand_dims(img,axis=0)
		result = model_CNN.predict(img)
		classname = ['Bánh bèo', 'Bánh bông lan',"Bánh chưng",'Bánh giò','Bánh mì','Bánh pía',
		'Bánh tráng nướng','Bánh xèo','Bún đậu','Nem chua']
		for i in range (0,10):
			if round(result[0][i])== 1:
				prediction = classname[i] 
				a = f"Đây là {prediction}"
				my_txt.configure(text=str(a))		
				if prediction=='Bánh bèo':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh bèo.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Bánh bông lan':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh bông lan.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Bánh chưng':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh chưng.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Bánh giò':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh giò.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Bánh mì':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh mì.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Bánh pía':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh pía.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Bánh tráng nướng':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh tráng nướng.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Bánh xèo':
					txt = open('D:\HKI-2022.2023\AI\Project\Bánh xèo.txt', 'r', encoding='UTF-8').read()
				elif prediction=='Nem chua':
					txt = open('D:\HKI-2022.2023\AI\Project\Bún.txt', 'r', encoding='UTF-8').read()
				
				else:
					txt = open('D:\HKI-2022.2023\AI\Project\Bún đậu.txt', 'r', encoding='UTF-8').read()
				
				en1.insert('end',txt)



img_choose= Image.open('D:\HKI-2022.2023\AI\Project\Pic.jpg')
img_choose = img_choose.resize((40,40))
img_choose=ImageTk.PhotoImage(img_choose)
button_choose=Button(frame,image=img_choose,font = ('Times',16),command = open_file)
button_choose.place(x=250,y=310)

img_check= Image.open('D:\HKI-2022.2023\AI\Project\check.png')
img_check = img_check.resize((40,40))
img_check=ImageTk.PhotoImage(img_check)
button_check=Button(frame,image=img_check,font = ('Times',16),command = check)
button_check.place(x=300,y=310)

img_cancel= Image.open('D:\HKI-2022.2023\AI\Project\Cancel.png')
img_cancel = img_cancel.resize((40,40))
img_cancel=ImageTk.PhotoImage(img_cancel)
button_cancel=Button(frame,image=img_cancel,font = ('Times',16),command = cancel)
button_cancel.place(x=350,y=310)



mainloop()
```
